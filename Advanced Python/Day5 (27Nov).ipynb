#### Plotting

- Univariate (Single axis)
- Bivariate (2D axis X,Y axis)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x = np.array([2,4,6,8])
y = x*2+3 #(7,11,15,19)
plt.plot(x,y,"", linewidth = 2,markersize = 10)
```

```python
markers = ["." , "," , "o" , "v" , "^" , "<", ">"]
colors = ['r','g','b','c','m', 'y', 'k']
x = np.arange(0,10)  #test x values.. every feature gets the same x values but you can generalize this
y = [s*x for s in np.arange(7)] #generate 7 arrays of y values 

for i in range(7): #for each of the 7 features 
    mi = markers[i] #marker for ith feature 
    xi = x #x array for ith feature .. here is where you would generalize different x for every feature
    yi = y[i] #y array for ith feature 
    ci = colors[i] #color for ith feature  
    plt.scatter(xi,yi,marker=mi, color=ci) 
plt.show() 
```

## Boxplot

- Univatiant

```python
plt.boxplot([1,10,20,200,50])
```

```python
import numpy as np
q1 = np.quantile([1,10,20,200,50],0.25)   # First half median
q2 = np.quantile([1,10,20,200,50],0.5)    # Median
q3 = np.quantile([1,10,20,200,50],0.75)   # Second half median
```

```python
q1, q2, q3
```

```python
IQR = 1.5 * (q3-q1)   # Inter Quartile Range
print(q1-IQR, q3+IQR)
```
