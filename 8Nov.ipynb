```python
import pandas as pd

df = pd.read_csv("olympics.csv", index_col = 0, skiprows = 1) # to read a particular csv file and to delete the row index and to skip how many rows

print(df.index) #shows only row index 
print(df.columns) #shows only columns

df
```

```python
for column in df.columns:
    if column[0:2] == "01":
        df.rename(columns={column:"Gold"+column[2:]}, inplace = True)
    if column[:2] == "02":
        df.rename(columns = {column:"Silver"+column[2:]}, inplace = True)
    if column[:2] == "03":
        df.rename(columns = {column:"Bronze"+column[2:]}, inplace = True)
    if column[:1] == "№":
        df.rename(columns = {column:"#"+column[1:]}, inplace = True)
        
df
```

```python
names = df.index.str.split("\s\(")
df.index = names.str[0]
df.index

```

```python
df["Id"] = names.str[1].str[:3] #to add a row with remaining words
df.drop("Totals", axis = 0, inplace=True)

df
```
