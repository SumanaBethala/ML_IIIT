```python
import pandas as pd
```

```python
# Creating a series data

a = pd.Series("Data", index = ["Index"])
a
```

```python
# If we are not assigning the index it will took the index value from 0 onwards

b = pd.Series(["Sam","is","Good"])
b
```

```python
# Using scalar values

c = [1,2,3,4,5,6]
c1 = pd.Series(c)
c1
```

```python
d = ["Apple", "Banana", "Cherry","Dates", "Emblic", "Fig"]

d1 = pd.Series(d,c)
d1
```

```python
# Data contains dictionary

e = {"Name":"Sumana", "Nickname": "Sumu"}
pd.Series(e)
```

```python
# In case that dictionary contains the list objects i.e., Ndarray

f = {"Name":["Sumana", "Shiny", "Sumanth", "Vasishta"], "Nickname": ["Sumu","Shine","Kothi","Vasi"]}
pd.Series(f)
```

```python
pd.DataFrame(a)
```

```python
# If we want to give the 2 dictionaries with same index value then we have to put them in list format 
# It will take default index value or else you can give your required index value 

g = {"Name":"Sumana", "Nickname": "Sumu", "Colour":"Blue"}
h = {"Name":"Swapna", "Nickname": "Sappu", "Colour":"Pink"}
s = pd.DataFrame([g,h])

# Or else put the indices in another dictionary

i = {"first":g, "Second":h}
pd.DataFrame(i)
```

```python
# A Dataframe can be set from the series format of pandas

data = {"first":b, "second":c1, "third":d1}
pd.DataFrame(data)
```

```python
j = [[1,2,3],[2,3,4]]
k = [["a","b","c"],["d","e","f"]]
pd.DataFrame([j,k], index = ["a","b"])
```

```python
j = [[1,2,3],[2,3,4]]
k = [["a","b","c"],["d","e","f"]]
data = {"first":j, "second":k}
pd.DataFrame(data)
```

## Reading a CSV (Comma Seperated Values) File

### Data Cleaning

* **Importing the Data set(CSV)**


* **Observe the Data set**

    * Identify the rows/columns and their relation
    
    * Identify the index column (Primary key column)
    
    * Check if there are any duplicate columns
    
    * which rows to be skipped
    
    
* **Assign the Index Column (to avoid creation of new index)**


* **Skip irrelevant rows (to avoid display of index row)**


* **Drop duplicate/irrelevant columns**

```python
import pandas as pd
pd.read_csv("pokemon.csv")
```

```python
# index_col indicates to which column is to be prferred as index instead of 0,1,2..

pd.read_csv("pokemon.csv",index_col = "Name", squeeze = True)
```

```python
# make the passed rows as heasder

pd.read_csv("pokemon.csv", header = [1,2])
```

```python
# usecols is used for to create a new dataframe by selecting the specified columns

pd.read_csv("pokemon.csv", usecols = ["Name", "Type 1", "Type 2", "Generation"])
```

```python
# Squeeze, it is an attribute if it True returns if there is only one column

#pd.read_csv("pokemon.csv", usecols = ["Name"])

pd.read_csv("pokemon.csv", usecols = ["Name"], squeeze = True)
```

```python
# skiprows used to skip the specified rows from the dataframe

pd.read_csv("pokemon.csv", skiprows = [1,3,4])
```
